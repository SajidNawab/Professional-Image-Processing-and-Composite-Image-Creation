**Q3:** In a post-apocalyptic world where Earth has become uninhabitable due to a nuclear fallout, a group of survivors is looking to you as their last hope. You are a skilled navigator on a mission to find the nearest potentially habitable planet to ensure the survival of humanity.

```python
spacecraft_position = np.array([500, 200, 100])
```

```python
planet_names = ["Proxima Centauri b", "Kepler-442b", "Kepler-186f", "TRAPPIST-1 Planets", "Gliese 581g", "Kepler-22b"]
```

```python
planet_coordinates = np.array([[1000, 800, 500],
                               [700, 300, 200],
                               [900, 600, 800],
                               [1200, 1000, 700],
                               [1500, 1200, 900],
                               [800, 500, 300]])
#print(np.array([1, 0, 0]))
```

```python
# Find the nearest habitable planet using the Euclidean distance formula
```

```python
##Calculate Distance of each Planet
ED=np.linalg.norm(planet_coordinates-spacecraft_position ,axis=0)
#ED
## Calculate Min Index
near_pI=np.argmin(ED)
#near_pI
##Find Name of nearest planet
name_P=planet_names[near_pI]
##Find  Coordinates of nearest planet
Cor_Value=planet_coordinates[near_pI]
##Display Name and Coordinates Value of nearest Planet
print(name_P,Cor_Value)
```

```python
# Now find the angle between the origin planet and each planet in 3d space using numpy functions
```

```python
##Calculate Vector of Planets
planet_V=planet_coordinates-spacecraft_position
planet_V
##Calculate Magitude of planet vector
planet_M=np.linalg.norm(planet_V , axis=1)
planet_M
# Calculate the dot product between the origin vector and each planet vector
dot_products = np.nansum(planet_V * np.array([1, 0, 0]), axis=1)

# Calculate the angle between the origin vector and each planet vector using the dot product
ang_R = np.arccos(dot_products / planet_M)

# Convert angles from radians to degrees
ang_R = np.degrees(ang_R)

# Display the angles between the origin and each planet
for i, angle in enumerate(ang_R):
    i=i+1
    print(f"The angle  {i} is {angle} degrees.")
    
```
